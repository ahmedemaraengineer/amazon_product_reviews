## Sentiment Analysis
* Sentiment analysis helps us make sense of unstructured text by automatically tagging it. Sentiment analysis helps us to process huge amounts of data in an efficient and cost-effective way.

```python
DATA_PATH = 'C:/Users/M/amazon_product_reviews/data/processed/amazon_reviews_processed.csv'
EXPORT_PATH = 'C:/Users/M/amazon_product_reviews/data/processed/amazon_reviews_processed_2.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import string
import re
import wordcloud
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

We will use two main classifiers:
* __Polarity__ : is float which lies in the range of [-1, 1] 
  1 for positive statement and -1 for negative statement 
* __Subjectivity__ : is used for individual sentences to determine whether a sentence expresses an opinion or not.

```python
from textblob import TextBlob
df['Text'] = df.Text.astype(str)

pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

df['polarity'] = df.Text.apply(pol)
df['subjectivity'] = df.Text.apply(sub)
df.head()
```

```python
df['Good_reviews']= df.Score
df.Good_reviews[df.Score <= 3] = 0
df.Good_reviews[df.Score >= 4] = 1
```

```python
df.polarity = df.polarity.round(2)
df.subjectivity = df.subjectivity.round(2)
```

```python
num_bins = 50
plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(df.polarity, num_bins, alpha=0.5)
plt.xlabel('polarity')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
sns.boxenplot(x='Good_reviews', y='polarity', data=df)
```

As we can see from this box plot,

* we have some good reviews that has very low polarity (very negative)
* Some bad reviews that has high polarity (positive statement)

#### Plotting the average of polarity and subjectivity over the scores through 1 to 5

```python
for i in range(1, 6):
    df2.loc[i, 'polarity'] = (df.head(100000).loc[df.Score == i].polarity.mean())
    df2.loc[i, 'subjectivity'] = (df.head(100000).loc[df.Score == i].subjectivity.mean())
```

```python
df2
```

```python
for index in range(1, 6):
    x = df2.polarity.loc[index]
    y = df2.subjectivity.loc[index]
    plt.scatter(x, y, color='blue')
    plt.text(x+0.001, y+0.001, index, fontsize=10)
    plt.xlim(-0.01, 0.4)
    plt.ylim(0.5, 0.6)
plt.title('Sentiment Analysis', fontsize=20)
plt.xlabel('<-- Negative -------- Positive -->', fontsize=15)
plt.ylabel('<-- Facts -------- Opinions -->', fontsize=15)

plt.show()
```

#### Plotting the average of polarity and subjectivity over the good reviews feature

```python
df3= pd.DataFrame()
for i in range (0,2):
    df3.loc[i,'polarity']= (df.head(100000).loc[df.Good_reviews == i].polarity.mean())
    df3.loc[i,'subjectivity']= (df.head(100000).loc[df.Good_reviews == i].subjectivity.mean())
df3
```

```python
df3['review']= df3.index
df3.review.replace({0:'Bad Reviews',1:'Good Reviews'}, inplace=True)
```

```python
df3
```

```python
plt.rcParams['figure.figsize'] = [10, 8]

for index, reviews in enumerate(df3.index):
    x = df3.polarity.loc[reviews]
    y = df3.subjectivity.loc[reviews]
    plt.scatter(x, y, color='blue')
    plt.text(x+.001, y+.001, df3['review'][index], fontsize=10)
    plt.xlim(-.01, .4) 
    plt.ylim(.5, .6) 
    
plt.title('Sentiment Analysis', fontsize=20)
plt.xlabel('<-- Negative -------- Positive -->', fontsize=15)
plt.ylabel('<-- Facts -------- Opinions -->', fontsize=15)

plt.show()
```

#### Punctuation vs Polarity

```python
plt.figure(figsize=(15, 8))
df3 = df.loc[df.upper <= 50]
sns.boxenplot(x='upper', y='polarity', data=df3)
plt.xlabel('Punctuation Quantity', fontsize=13)
plt.ylabel('Polarity Score', fontsize=13)
plt.title('Punctuation vs Polarity')
plt.show()
```

We can see that when the value of punctuation is low, polarity is higher
this is because people who are paying attention to punctuation tend to be more realistic about their evaluation

### Examining Subjectivity 
* Subjectivity is used for individual sentences to determine whether a sentence expresses an opinion or not.
* ubjectivity is also a float that lies in the range of [0,1]
* When it is close to 0, it is more about facts
* When subjectivity increases, it comes close to be an opinion

```python
plt.figure(figsize=(10,5))
sns.distplot(df['subjectivity'], hist=True, kde=True, 
             bins=int(30), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
plt.xlim([-0.001,1.001])
plt.xlabel('Subjectivity', fontsize=13)
plt.ylabel('Frequency', fontsize=13)
plt.title('Distribution of Subjectivity Score', fontsize=15)
```
